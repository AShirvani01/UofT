3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import pymc as pm
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import arviz as az
```

```python
n = 100
true_theta, true_tau = 0,1
x = stats.norm(true_theta, true_tau**(-0.5)).rvs(n)
```

### $\theta \sim N(\mu=0,\tau_0=1), \tau \sim Gamma(\alpha=1/2, \beta=1/2)$

```python
conjugate_normal_gamma = pm.Model()
with conjugate_normal_gamma:

    theta = pm.Normal("theta", mu=0, tau=1)  
    tau = pm.Gamma("tau", alpha=1/2, beta=1/2)

    # Likelihood (sampling distribution) of observations
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
```

```python
m = 10000
with conjugate_normal_gamma:
    idata = pm.sample(draws=m, chains=4, tune=100)
```

```python
az.plot_trace(idata, combined=True)
plt.tight_layout()
```

```python
az.plot_forest(idata, var_names=["theta", "tau"], 
               combined=True, hdi_prob=0.95, r_hat=True)
```

```python
az.summary(idata, round_to=5)
```

### $\theta \sim LogNormal(\mu=0,\tau_0=1), \tau \sim Exp(\lambda=1)$

```python
lognorm_exp = pm.Model()
with lognorm_exp:

    theta = pm.LogNormal("theta", mu=0, tau=1)  
    tau = pm.Exponential("tau", lam=1)

    # Likelihood (sampling distribution) of observations
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
```

```python
m = 10000
with lognorm_exp:
    idata = pm.sample(draws=m, chains=4, tune=100)
```

## Question 3

### Slice Sampling Algorithm

- Initialized at $x = 0.25$

#### slice_f_at_y(f, x, y, x_grid) 
- finds an interval where f(x) > y
- assign interval bounds to a and b
- sample $\tilde x$ from Uniform(a,b)
- if $f(\tilde x) > y$, then accept $\tilde x$
- if not acccepted, it narrows the interval and calls slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1)

#### slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1)
- Iteratively resample $\tilde x$ and narrow the interval until $f(\tilde x) > y$, then accept

### Slice Sampling within Gibbs instead of Metropolis-Hastings

- Initialize $\tau_0, \theta_0$ to some initial values

#### Sampling $\tau | \theta,x$

- Sample slice height,$y$, from $Uniform(0,f(\tau_t | \theta,x))$ using the previous $\theta$ value (if first iteration, use $\theta_0$)
- Find interval [a,b] where $f(\tau|\theta,x) > y$
- If can't find interval, keep reducing interval until $f(\tau|\theta,x) > y$, then accept sample

#### Sampling $\theta | \tau,x$
- Same procedure as $\tau | \theta,x$, but using the previous sampled $\tau$ in the conditional up to a normalization constant

---

- Repeat iteratively for desired number of iterations (ideally enough for MC to converge)
- The curves that we are sampling beneath in each step are the full conditional distributions up to a normalization constant (which in this case are both $f \sim \beta(1.5,3)$

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = 0.02
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-0.02,0.02]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')
    
plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```
